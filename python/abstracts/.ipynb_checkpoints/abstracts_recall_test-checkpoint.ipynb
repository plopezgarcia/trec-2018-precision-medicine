```python
topics_all = evaluation.get_topics('./topics/topics2017.xml')
qrels_all = evaluation.get_qrels('./gold-standard/abstracts.2017.qrels')
TOPIC = 2
```

```python
topic = topics_all[(topics_all.topic==TOPIC)]
qrels = evaluation.qrels_of_topics(qrels_all, topic)
topic
```

```python
run, params = evaluation.run(topic)
params['query_template'] = 'topic-optimized.json'
results, aggregated = evaluation.evaluate(qrels[(qrels.relev.isin([2]))], run)
run, params = evaluation.run(topic, params)
aggregated
```
